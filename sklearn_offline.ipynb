This notebook was made to test out the model locally before training and deploying it on SageMaker

```
# imports
import pandas as pd
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
```

Creation of model:

```
# Read in csv training file
train_data = pd.read_csv("plagiarism_data/train.csv", header=None, names=None)

# Labels are in the first column
train_y = train_data.iloc[:,0]
train_x = train_data.iloc[:,1:]

# define model
# model = svm.SVC(kernel='poly', C=1, gamma=0.01)
# model = svm.SVC()
# model = svm.LinearSVC()
# model = svm.SVC(kernel='linear', C=1, gamma=0.001)
# model = GaussianNB()
# model.fit(train_x, train_y)
model = svm.LinearSVC()

# train model
model.fit(train_x, train_y)

```

Prediction:

```
test_data = pd.read_csv("plagiarism_data/test.csv", header=None, names=None)
# labels are in the first column
test_y = test_data.iloc[:,0]
test_x = test_data.iloc[:,1:]
pred = model.predict(test_x)

for value in zip(pred, test_y.values):
    print(value)
# print("Predictions: ", pred)
# print("Actual: ", test_y.values)
```

Accuracy:

```
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, test_y)
print("Accuracy: ", acc)
```
